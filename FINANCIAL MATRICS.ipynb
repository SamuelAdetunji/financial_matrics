```python
import matplotlib.pyplot as plt
import numpy as np
import math

intro = 'PROVIDE THE FOLLOWING DATAS TO CALCULATE \n'        #TITLES TO BE DISPLAY
ingra = 'THIS IS THE ANALYSIS AND GRAPHICAL PRESENTATION OF THE VALUSES ENTER \n'
width = 100                             #TITLE ALLIANMENT
print(intro.center(width))

def input_handling(valid):
    while True:
        try:
            values = int(input(valid))
            return values
        except ValueError:
             print("INVALID INPUT. Please enter a valid numeric value e.g 12546.")
TE = input_handling("Provide TOTAL EXPENSES $: \n")
TR = input_handling("Provide TOTAL REVENUE $: \n")        #INPUT BODY WITH ERROR HANDLING FUNCTION
N_profit = input_handling("Provide NET PROFIT $: \n")
G_profit = input_handling("Provide GROSS PROFIT $: \n")                 
Ope_profit = input_handling("Provide TOTAL OPERATING PROFIT $: \n")
N_profit_invest = input_handling("Provide NET PROFIT INVESTMENT $: \n")
Co_of_invest = input_handling("Provide COST OF INVESTMENT $: \n")
N_earning = input_handling("Provide NET EARNNG $: \n")
Nu_out_standshare = input_handling("Provide NUMBER OF OUT STANDING SHARES : \n")
print( )
threshold = 25

def main():                            #MAIN FUNCTION THAT HOST THREE DIFFERENT FUNCTIONS
    def finance(TE, TR, N_profit, G_profit, Ope_profit, N_profit_invest, Co_of_invest, N_earning, Nu_out_standshare):
        NI = (TR - TE)
        if TR > TE:
            print("PROFIT OF ",NI," IS MADE \n")
        else:                                              
            print("LOSS OF ",NI, " IS MADE \n")
        
        GPM = (G_profit / TR) * 100
        GPM_V = round(GPM-threshold,2)
        if GPM > threshold:
            print("HIGH GROSS PROFIT OF ",GPM_V, "IS GENETATED \n")
        else:
            print("GROSS PROFIT MARGIN OF",GPM_V, "IS NOT CONSIDER HIGH \n")
                                                                                 #FUNCTION 1 BODY AND ITS STATISTICS
        NPM = (N_profit / TR) * 100
        NPM_V = round(NPM-threshold,2)
        if NPM > threshold:
            print("HIGH NET PROFIT MARGIN OF ",NPM_V, "IS GENERATED \n")
        else:
            print("NET PROFIT MARGIN OF ",NPM_V, "IS NOT CONSIDER HIGH \n")
        
        OPM = (Ope_profit / TR) * 100
        OPM_V = round(OPM-threshold,2)
        if OPM > threshold:
            print("HIGH OPERATING PROFIT MARGIN OF ",OPM_V, "IS GENERATED \n")
        else:
            print("OPERATING PROFIT MARGIN OF ",OPM_V, "IS NOT CONSIDER HIGH \n")
        
        ROI = ((N_profit_invest - Co_of_invest) / Co_of_invest) * 100
        if ROI > 0:
            print("ROI IS POSITIVE. THE INVESTMENT IS PROFITABLE \n")
        elif ROI < 0:
            print("ROI IS NEGATIVE. THE INVESTMENT RESULTED IN A LOSS \n")
        else:
            print("ROI IS ZERO. THE INVESTMENT IS NEIGHER GAIN NOR LOST \n")
    
        EPS = (N_earning / Nu_out_standshare)
        if EPS > 0:
            print("EPS IS POSITIVE. THE INVESTMENT IS PROFITABLE \n")
        elif EPS < 0:
            print("EPS IS NEGATIVE. THE INVESTMENT RESULTED IN A LOSS \n")
        else:
            print("EPS IS ZERO. THE INVESTMENT IS NEIGHER GAIN NOR LOST \n")
        
        values = [NI, GPM, NPM, OPM, ROI, EPS, TE, TR]
        plot_value = [math.ceil(val) for val in values]    # FUNCTION 1 LOOPING AND RETURN 
        return plot_value

    def scale(x):
        x_side = np.arange(0,x,10)         #FUNCTION 2 BODY FOR GRAPH SCALES x
        return x_side

    print(ingra.center(width))
    def p_graph():
        plt.plot(x_axis, y_axis, lw=2, marker='o', ms=7, mfc='r', mec='k')
        plt.xlabel("Metrics")               #FUNCTION 3 FOR THE GRAPH PLOTTING
        plt.ylabel("Values")
        plt.xticks(x_axis, ['NI', 'GPM', 'NPM', 'OPM', 'ROI', 'EPS', 'TE', 'TR'])
        for x_ax, y_ax in zip(x_axis,y_axis):
            plt.text(x_ax, y_ax, str(y_ax))
        plt.grid(lw=2, color='g')
        title_text="Financial Metrics"
        plt.title(title_text)
        plt.show()
        
    y_axis = finance(TE, TR, N_profit, G_profit, Ope_profit, N_profit_invest, Co_of_invest, N_earning, Nu_out_standshare)
    x_axis = scale(80)
    graph = p_graph()                                 #CALLING THE FUNCTION 1,2,3 INSIDE THE THEIR SCOPES
    
    return y_axis, x_axis, graph     #RETURNING THE VARIABLE USED TO CALL FUNCTION 1,2,3 INSIDE THE MAIN FUNCTIN SCOPES

y_axis, x_axis, graph = main()      #USING UNPACK STYLE TO ASIGHN THE THREE VARIABLE 
                                    #TO THE MAIN FUNCTION FOR EASY ACCESS BY EACH           
```
